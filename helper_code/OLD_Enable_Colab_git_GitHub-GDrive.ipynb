###This code is to connect to a git-enabled Colab session <br /> **before** ***or*** **after** you have already cloned your repo to Google Drive.

1) initialization of a new GitHub repo in your "local" Google Drive, and git connection information enabled for Colab to mediate between the two

2) after you have already cloned the repo to your Google Drive, but have 
been disconnected from Colab runtime and need to refresh your Colab's git-enabled connection betwen GitHub and your local Google Drive cloned repo


```
# Mount your personal Google Drive in Colab at "/content/drive"
from google.colab import drive
drive.mount('/content/drive')
```

```
# repository path setup

# Add your desired path into which you will clone the repository
GDRIVE_PATH = "/content/drive/My Drive/Colab Notebooks/NRUHSE_2_Kaggle_Coursera/final"

# Enter your token here (get it from github.com/settings/tokens -- it should be unique to you, and of course change the user name and email)
GIT_TOKEN = "123456789123456789"
GIT_USERNAME = "migai"
GIT_REPO = "Kag"

GITHUB_REPO_PATH = "https://" + GIT_TOKEN + "@github.com/" + GIT_USERNAME + "/" + GIT_REPO + ".git"
GDRIVE_REPO_PATH = GDRIVE_PATH + "/" + GIT_REPO
```

```
# Define function to quickly push changes to GitHub repo

def git_push(push_message, gdrive_path=GDRIVE_REPO_PATH, github_path=GITHUB_REPO_PATH):

  %cd "{gdrive_path}"
    
  !git config user.email "gaidis@alum.mit.edu"
  !git config user.name "migai"

  !git add .
  
  !git commit -m "{push_message}"
  
  !git remote remove origin
  !git remote add origin "{github_path}"

  !git push origin master
  return
```
